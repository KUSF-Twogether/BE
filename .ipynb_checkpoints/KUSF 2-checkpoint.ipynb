```python
def hello(A : str, B : str):
    '''A가 B에게 인사를 하는 함수 만들기
    Args:
        A(str) : 인사를 건네는 사람
        B(str) : 인사를 받는 사람   
    '''
    print(f"{A}가 {B}에게 인사를 합니다.")
hello("김태우","Messi")
```

```python
score = 0
while True :
    
    order = input("실행할 명령번호를 입력하시오.")

    if order == "1" :
        print("먹이를 줍니다")
        score += 5
    elif order == "2" :
        print("놀아 줍니다")
        score += 7
    elif order == "3" :
        print("산책을 시킵니다")
        score += 10
    elif order == "4" :
        print("목욕을 시킵니다")
        score += 3
    elif order == "q" :
        print(f'강아지의 현재 행복지수는 {score}점입니다.')
        break
```

```python
class Poketmonster:
    def __init__(self,name, skill):
        self.name = name
        self.skill = skill
        
    def introduce(self):
        print(f"제 이름은 {self.name}이고, 기술은 {self.skill}이(가) 있습니다.")
        
class Pikachu(Poketmonster):
    def electricity(self):
        print("저는 제 기술 중 백만볼트를 사용합니다.")
        
class Squirtle(Poketmonster):
    def water(self):
        print("저는 제 기술 중 아쿠아테일을 사용합니다.")
        
p1 = Pikachu("피카츄", "백만볼트")
p2 = Squirtle("꼬부기", "아쿠아테일")

p1. introduce()
p1. electricity()

p2. introduce()
p2. water()
```

```python
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def eat(self):
        print("밥을 먹습니다")
        
    def introduce(self) :
        print(f"제 이름은 {self.name}이고, 나이는 {self.age}입니다.")
        
p1 = person("김태우", 24)
p2 = person("필 포든", 24)

p1.eat()
p1.introduce()

p2.eat()
p2.introduce()
```

```python
class SmartPhone():
    def __init__(self, manufacturer: str, price: int):
        self.manufacturer = manufacturer
        self.price = price
    
    def call(self):
        print(f"{self.manufacturer} 스마트폰으로 전화통화를 합니다.")

    def web_surf(self):
        print("웹서핑을 합니다.")

    def play_game(self):
        print(f"{self.price}원 스마트폰으로 게임을 합니다.")


smartphone1 = SmartPhone("애플", 1400000)
smartphone2 = SmartPhone("삼성", 1200000)

smartphone1.call() 
smartphone1.web_surf()

smartphone2.call() 
smartphone2.play_game()
```

```python
class Iphone(SmartPhone):
    def feeling(self):
        print("감성을 충족시켜줍니다.")
    
phone1 = Iphone("애플", 1400000)
    
phone1.call()
phone1.play_game()
phone1.feeling()
```

```python

class Calc:
    @staticmethod
    def add(x, y):
        return x + y
    
    @staticmethod
    def sub(x, y):
        return x - y
    
    @staticmethod
    def mul(x, y):
        return x * y
    
    @staticmethod
    def div(x, y):
        return x / y
print(Clac.add(5, 7))
print(Clac.sub(8, 3))
print(Clac.mul(4, 6))
print(Clac.div(9, 2))
```

```python
class ManCity:
    mancity_cnt = 0
    
    def __init__(self, name, backnum, pos):
        self.name = name
        self.backnum = backnum
        self.pos = pos
        ManCity.mancity_cnt += 1
        
    def team(self):
        print(f"저는 맨체스터 시티 소속입니다.")
        
class Player(ManCity):
    def player(self):
        print(f"제 이름은 {self.name}이고, 저의 등번호는 {self.backnum}번이며, 저의 포지션은 {self.pos}에서 주로 뜁니다.")

p1 = Player("필 포든", 47, "LW")
p2 = Player("케빈 더브라위너", 17, "CAM")
p3 = Player("로드리", 16, "CDM")
p4 = Player("에데르송", 1, "GK")

p1.player()
p1.team()

p2.player()
p2.team()

p3.player()
p3.team()

p4.player()
p4.team()

print("=====================================")
print(f"현재까지 라인업 등록 인원 {ManCity.mancity_cnt}명.등록 가능 인원은 {11-ManCity.mancity_cnt}명 입니다.")
```
